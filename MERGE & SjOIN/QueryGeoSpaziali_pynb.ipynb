```python
%matplotlib inline

import pandas as pd
import geopandas
```

# METTIAMO I GEODATAFRAME E CREAMO PROVINCE2,COMUNI2 E REGIONI2, TUTTE E 3 CON I NOMI E LE GEOMETRIE.

```python
province = geopandas.read_file('/content/drive/MyDrive/Colab Notebooks/Limiti01012021_g/ProvCM01012021_g/ProvCM01012021_g_WGS84.dbf')
comuni = geopandas.read_file('/content/drive/MyDrive/Colab Notebooks/Limiti01012021_g/Com01012021_g/Com01012021_g_WGS84.dbf')
regioni = geopandas.read_file('/content/drive/MyDrive/Colab Notebooks/Limiti01012021_g/Reg01012021_g/Reg01012021_g_WGS84.dbf')
```

```python
comuni
```

```python
comuni2 = comuni[['COMUNE','geometry']]
comuni2
```

```python
province
```

```python
province2 = province[['DEN_UTS','geometry']]
province2
```

```python
regioni
```

```python
regioni2 = regioni[['DEN_REG','geometry']]
regioni2
```

# **PRIMO APPROCCIO A SJOIN**

```python
ComuniProvince = geopandas.sjoin(comuni2,province2,op = 'within',how='left')  #op: tipo di operazione sulle geometrie,come intersects, touches e altri.
ComuniProvince
```

# **ESERCIZIO CON TUTTO**

```python
reg_Prov = geopandas.sjoin(province2,regioni2,op='within',how='left')  #sjoin funziona con left,inner, ma non con right e outer.
reg_Prov.plot()
```

```python
reg_Prov
```

```python
Nprov = reg_Prov.groupby('DEN_REG')[['DEN_UTS']].count().reset_index()
Nprov
```

**COME FACCIO A COLORARE LE REGIONI SULLA BASE DEL NUMERO DELLE REGIONI?**

**possiamo fare sjoin se sono 2 geodataframe, merge se sono 1 dataframe e un geodf, o 2 dataframe.**

```python
reg_Prov = pd.merge(regioni2,Nprov,how ='left',on = 'DEN_REG')
```

```python
reg_Prov.plot(column ='DEN_UTS',legend = True,figsize=(10,10))
```

**ES SOPRA COMPRESSO**

```python
reg_Prov = geopandas.sjoin(province2,regioni2,op='within',how='left')
Nprov = reg_Prov.groupby('DEN_REG')[['DEN_UTS']].count().reset_index()
reg_Prov = pd.merge(regioni2,Nprov,how ='left',on = 'DEN_REG')
reg_Prov.plot(column ='DEN_UTS',legend = True,figsize=(10,10))
```

# **ES2**

```python
#trova per ogni comune la provincia in cui si trova quel comune
#effettuando una joinSpaziale(sjoin):prende la geometria del comune
#e cerca in quale geometria della provincia si trova.
#restuisce im dataframe

Com_Prov = geopandas.sjoin(comuni2,province2,op = 'within',how='left')
#_______________________________________________________________________________ _ _ _ _ _ _  _  _  _   _  _  _   _    _    _ 


#raggruppa i comuni di ogni provincia e li conta.
#restituisce un dataframe

NCom = Com_Prov.groupby('DEN_UTS')['COMUNE'].count().reset_index()
#_______________________________________________________________________________ _ _ _ _ _ _  _  _  _   _  _  _   _    _    _ 

#unisce il geodataframe delle province con il dataframe che contiene il numero dei comuni per ogni provincia.
#restituisce un geodataframe che è uguale a province2 ma con una colonna in più,che è il numeo di comuni per ogni provincia.

Comuni_Province = pd.merge(province2,NCom,how ='left',on = 'DEN_UTS')
#_______________________________________________________________________________ _ _ _ _ _ _  _  _  _   _  _  _   _    _    _ 

#visualizza le province colorate in base alla densità dei comuni(column ='COMUNE')

Comuni_Province.plot(column ='COMUNE',legend = True,figsize=(10,10))
#_______________________________________________________________________________ _ _ _ _ _ _  _  _  _   _  _  _   _    _    _ 
```
