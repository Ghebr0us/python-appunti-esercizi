```
import contextily as ctx
```

```
ax = Piste_Ciclabili.to_crs(epsg=3857).plot(figsize=(15,10))
ax1 = Quartieri_Milano.to_crs(epsg=3857).plot(ax=ax,facecolor='none',edgecolor='k')
ctx.add_basemap(ax)
```

```
# trovare il quartiere meno esteso
Minus = Quartieri_Milano[Quartieri_Milano['Shape_Area'] == min(Quartieri_Milano['Shape_Area'])]
Minus
```

```
# trovare il quartiere più esteso
Maxi= Quartieri_Milano[Quartieri_Milano['Shape_Area'] == max(Quartieri_Milano['Shape_Area'])]
Maxi
```

```
# visualizzare i due quartieri sulla mappa di Milano con i colori indicati nella figura sotto
ax = Maxi.to_crs(epsg=3857).plot(facecolor='y',edgecolor='k',figsize=(10,10),alpha=0.5)
Minus.to_crs(epsg=3857).plot(facecolor = 'red',alpha = 0.5,ax=ax)
ctx.add_basemap(ax)
```

```
# salvare in un geodataframe le informazioni sul quartiere di Affori, prendendole dal geodataframe dei quartieri
```

```
Affori = Quartieri_Milano[Quartieri_Milano.NIL == 'AFFORI']
Affori
```

```
# trovare le piste ciclabili che passano per il quartiere di Affori
```

```
Piste_Affori = Piste_Ciclabili[Piste_Ciclabili.intersects(Affori.geometry.squeeze())]
Piste_Affori
```

```
# visualizzare su una mappa il quartiere di Affori e le piste ciclabili che lo attraversano
# evidenziare bene in rosso le piste ciclabili come nell'esempio sotto
```

```
ax = Affori.to_crs(epsg=3857).plot(facecolor='none',figsize=(15,10))
Piste_Affori.to_crs(epsg=3857).plot(ax=ax,color='r')
ctx.add_basemap(ax)
```
