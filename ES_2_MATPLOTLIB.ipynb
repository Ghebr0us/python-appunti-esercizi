```python
import pandas as pd  
import matplotlib.pyplot as plt
```

```python
data = {'last_name': ['Boyer', 'Copeland', 'Daniel', 'Serrano','Terrel','Vargas'],
        'quantity': [766,213,812,269,100,114],
        'total_amount':[643049.37,203549.39,744259.20,253183.68,93851.95,123156.90]}  
  
# Create DataFrame  
df = pd.DataFrame(data)  
  
# Print the output.  
df
```

**b)**visualizzi due grafici a barre affiancati che indichino quantit√† vendute e valore venduto per ogni
dipendente. Il risultato deve essere questo:

```python
fig,(ax1,ax2) = plt.subplots(1,2)
fig.suptitle('')
fig.set_size_inches(15,8)
plt.rcParams.update({'font.size':30})
fig.autofmt_xdate(rotation='vertical')
ax1.bar(df['last_name'],df['quantity'])
ax2.bar(df['last_name'],df['total_amount'])
```

**c)**visualizzi i grafici precedenti con un titolo, gli assi e la griglia.


```python
fig,(ax1,ax2) = plt.subplots(1,2)  #unisco i 2 grafici
fig.suptitle('GRUPPO DI GRAFICI')  #-----------|
fig.set_size_inches(15,8)                     #|-grafica e ottimizzazione
plt.rcParams.update({'font.size':10}          #|
fig.autofmt_xdate(rotation='vertical'))#-------|
ax1.bar(df['last_name'],df['quantity'])  #creo primo grafico
ax2.bar(df['last_name'],df['total_amount'])#creo secondo grafico
ax1.grid()
ax2.grid()  #griglia
```
