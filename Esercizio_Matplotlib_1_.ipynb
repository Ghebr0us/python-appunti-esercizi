#ESERCIZIO 1

**a)**crei un dataframe contenente le informazioni riportate nella tabella sottostante, relativa al numero di
ordini (order_numbers) seguiti da vari commessi (last_name) di un negozio.

```python
data = {'last_name': ['Boyer', 'Copeland', 'Daniel', 'Serrano','Terrel','Vargas'],
        'orders_number': [185, 50, 196, 58,24,28]}  
  
# Create DataFrame  
df = pd.DataFrame(data)  
  
# Print the output.  
df
```

visualizzi un grafico a barre verticali per rappresentare i dat. Il risultato deve essere questo:

```python
data = list(df['orders_number'])
names = list(df['last_name'])
fig = plt.figure()
ax = plt.axes()
ax.bar(names, data)


plt.show()
```


**c)**visualizzi lo stesso grafico ma con altri colori delle barre, un titolo appropriato al grafico e i nomi degli
assi.

```python
data = list(df['orders_number'])
names = list(df['last_name'])
fig = plt.figure(figsize = (10,10))
ax = plt.axes()
ax.bar(names, data, color = 'r')

ax.set_xlabel('last_name')
ax.set_ylabel('orders_number')
fig.suptitle('Istogramma')
plt.show()
```

**d)**Visualizzare lo stesso grafico ma con le barre in ordine decrescente

```python
fig = plt.figure(figsize = (8,7))
ax = plt.axes()
ax.set_xlabel('last_name', fontsize = 20)
ax.set_ylabel('orders_number', fontsize = 20)
fig.suptitle('ISTOGRAMMA', fontsize = 30)

ax.bar(df.sort_values("orders_number", ascending = False)["last_name"], df["orders_number"].sort_values(ascending = False), color = "c")
plt.show()
```

**e)**visualizzi un grafico a barre orizzontali per rappresentare i dat. Il risultato deve essere uguale a questo:

```python
fig = plt.figure()
ax = plt.axes()
ax.barh(names, data)


plt.show()
```

**f)**crei un nuovo dataframe contenente la percentuale di ordini consegnati seguiti da ogni componente
dello staff. Visualizzare il cognome del componente dello staff e la percentuale di ordini. Ordinare in
ordine alfabetico sul cognome del componente dello staff. Il risultato deve essere questo:

```python
data = {'last_name': ['Copeland', 'Serrano', 'Boyer', 'Daniel','Vargas','Terrel'],
        'orders_perc': [9.242144, 10.720887, 34.195933, 36.229205,5.175601,4.436229]}  
  
# Create DataFrame  
df = pd.DataFrame(data)  
  
# Print the output.  
df
```

**g)**visualizzi un grafico a torta per rappresentare i dati estratti. Il risultato deve essere questo:

```python
perc = df['orders_perc']

plt.rcParams.update({'font.size':14})

fig = plt.figure(figsize = [10,10])
ax = plt.axes()

ax.pie(perc, labels=df['last_name'],autopct='%1.2f%%')

plt.show()
```

```python
cols=['c','m','r','g','r','m']
perc = df['orders_perc']

plt.rcParams.update({'font.size':14})


fig = plt.figure(figsize = [10,10])
ax = plt.axes()

ax.pie(perc, labels=df['last_name'],colors=cols,autopct='%1.2f%%')

fig.suptitle('Grafico a Torta')
plt.show()
```

```python
size = list(df['orders_perc'])
label =  list(df['last_name'])
squarify.plot(sizes=size, label=label, alpha=0.6 )
plt.axis('off')
plt.show()
```
